```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Mock financial data
investment_data = {
    'year': [2020, 2021, 2022, 2023],
    'stocks': [7, 10, -5, 15],
    'bonds': [2, 3, 1, 2],
    'real_estate': [3, 4, 3, 5]
}

financial_df = pd.DataFrame(investment_data)

# User data
user_data = {
    'initial_savings': 50000,
    'monthly_contribution': 500,
    'current_age': 30,
    'retirement_age': 65,
    'health_status': 'good',
    'risk_tolerance': 'medium'  # Options: low, medium, high
}

# Convert user data into a DataFrame for consistency
user_df = pd.DataFrame([user_data])

# Prepare data for the model
X = financial_df['year'].values.reshape(-1, 1)
y_stocks = financial_df['stocks'].values
y_bonds = financial_df['bonds'].values
y_real_estate = financial_df['real_estate'].values

# Train models for each investment type
model_stocks = LinearRegression().fit(X, y_stocks)
model_bonds = LinearRegression().fit(X, y_bonds)
model_real_estate = LinearRegression().fit(X, y_real_estate)

# Predict future returns (for example, next 10 years)
future_years = np.array([2024 + i for i in range(10)]).reshape(-1, 1)
predicted_stocks = model_stocks.predict(future_years)
predicted_bonds = model_bonds.predict(future_years)
predicted_real_estate = model_real_estate.predict(future_years)

# Combine predictions into a DataFrame
predictions_df = pd.DataFrame({
    'year': future_years.flatten(),
    'stocks': predicted_stocks,
    'bonds': predicted_bonds,
    'real_estate': predicted_real_estate
})

# Define allocation based on risk tolerance
def allocate_investments(risk_tolerance):
    if risk_tolerance == 'low':
        return {'stocks': 0.2, 'bonds': 0.5, 'real_estate': 0.3}
    elif risk_tolerance == 'medium':
        return {'stocks': 0.4, 'bonds': 0.4, 'real_estate': 0.2}
    else:  # high risk tolerance
        return {'stocks': 0.6, 'bonds': 0.2, 'real_estate': 0.2}

# Get allocation for the user
allocation = allocate_investments(user_data['risk_tolerance'])

# Calculate investment growth over time
def calculate_investment_growth(user_data, predictions_df, allocation):
    years_to_retirement = user_data['retirement_age'] - user_data['current_age']
    total_years = min(years_to_retirement, len(predictions_df))
    
    investment_values = {
        'year': [],
        'total_value': [],
        'stocks_value': [],
        'bonds_value': [],
        'real_estate_value': []
    }

    current_savings = user_data['initial_savings']
    monthly_contribution = user_data['monthly_contribution']
    
    for i in range(total_years):
        year = predictions_df['year'].iloc[i]
        stock_return = predictions_df['stocks'].iloc[i] / 100
        bond_return = predictions_df['bonds'].iloc[i] / 100
        real_estate_return = predictions_df['real_estate'].iloc[i] / 100

        # Contributions over the year
        current_savings += monthly_contribution * 12
        
        # Calculate values based on allocations
        stocks_value = current_savings * allocation['stocks'] * (1 + stock_return)
        bonds_value = current_savings * allocation['bonds'] * (1 + bond_return)
        real_estate_value = current_savings * allocation['real_estate'] * (1 + real_estate_return)
        
        total_value = stocks_value + bonds_value + real_estate_value
        
        # Update savings
        current_savings = total_value
        
        # Store the values
        investment_values['year'].append(year)
        investment_values['total_value'].append(total_value)
        investment_values['stocks_value'].append(stocks_value)
        investment_values['bonds_value'].append(bonds_value)
        investment_values['real_estate_value'].append(real_estate_value)
    
    return pd.DataFrame(investment_values)

investment_growth_df = calculate_investment_growth(user_data, predictions_df, allocation)

# Plot the investment growth
plt.figure(figsize=(10, 6))
plt.plot(investment_growth_df['year'], investment_growth_df['total_value'], label='Total Value')
plt.plot(investment_growth_df['year'], investment_growth_df['stocks_value'], label='Stocks Value')
plt.plot(investment_growth_df['year'], investment_growth_df['bonds_value'], label='Bonds Value')
plt.plot(investment_growth_df['year'], investment_growth_df['real_estate_value'], label='Real Estate Value')

plt.xlabel('Year')
plt.ylabel('Value ($)')
plt.title('Investment Growth Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
